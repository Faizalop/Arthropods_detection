```python
from collections import Counter
```

```python
object_prediction_list = result.object_prediction_list
```

```python
# object_prediction_list
```

```python
class_names = [prediction.category.name for prediction in object_prediction_list]
```

```python
# Count the occurrences of each class
class_counts = Counter(class_names)

# Print the counts
for class_name, count in class_counts.items():
    print(f"{class_name}: {count}")
```

```python
class_map = {
    'MA': 'Melon Aphid',
    'NT': 'Nesidiocorius Tenius',
    'OI': 'Orius Insidiosus',
    'WFT': 'Western Flower Thrips',
    'WF': 'White Fly',
    'TS': 'Two Spotted Spidermite'
}
```

```python
import pandas as pd
data = {
    'Abbreviation': [],
    'Full Name': [],
    'Count': []
}

for class_name, count in class_counts.items():
    data['Abbreviation'].append(class_name)
    data['Full Name'].append(class_map.get(class_name, class_name))
    data['Count'].append(count)

df = pd.DataFrame(data)

print(df)
```
